# Demo for ppn run analysis

```python
%pylab nbagg
from nugridpy import ppn
from nugridpy import utils
```

`data_dir` points to a directory with a PPN run, for example:

```python
data_dir="/data/nugrid_vos/data/projects/ppn/examples/ppn_iprocess_1zone"
```

```python
pa = ppn.abu_vector(data_dir)
px = ppn.xtime(data_dir)
```

```python
pa.get(560,decayed=True)
```

```python
px.cols
```

```python
from nugridpy import utils as ut
```

```python
ut.linestylecb(4)
```

```python
figure(155)
things=['NI 66',"ZN 72","GA 75","GE 78","AS 79","SE 84"]
for i,thing in enumerate(things):
    lstyle=ut.linestylecb(i)[0]
    #px.plot('cycle',thing,shape=lstyle,markevery=ut.linestylecb(i)[3])
    m=10*i+2*i
    px.plot('cycle',thing,shape=lstyle,legend=thing)
title("");xlabel('cycle');ylabel('mass fraction')
xlim(475,600);legend(loc=0)
```

```python
pa.abu_chart(560, plotaxis=[1,55,1,42],  color_map='viridis',imlabel_fontsize=10,\
             ilabel=True, imlabel=False)
```

```python
close(5); figure(5)
pa.iso_abund(520,decayed=True,amass_range=(52,90))
```

```python
close(6); figure(6)
pa.iso_abund(520,decayed=False,amass_range=(52,90))
```

```python
pa.elemental_abund?
```

```python
figure(156)
pa.elemental_abund(520,zrange=[10,65],ref=0, marker='o',linestyle='dotted',\
                   title_items=['densn','mod'],ylim=[-10,-2])
```
