```python
%pylab 

import sys
sys.path.insert(1,"/user/scratch14_wendi3/NuGrid/NuGridPy")

from nugridpy2 import nugridse as mp
from nugridpy2 import mesa as ms


# default data location (try mirror if default
# is not available)
#data_dir="/tmp/nugrid_data"   

# mirror data locations (some mirrors 
# may not be available at some time)

data_dir="/data/nugrid_cedar"  # may be default    
# data_dir="/data/nugrid_rpod4"    
#data_dir="/data/nugrid_vos"    


ms.set_nugrid_path(data_dir)
mp.set_nugrid_path(data_dir)
```

```python
fsize=16
engenlevels=[1e2,1e4,1e6,1e8,1e10,1e12,1e14]
res=10000
```

### Figure 15

```python
s1=ms.history_data(mass=1.0,Z=0.0001)
```

```python
ifig=107;close(ifig);figure(ifig,figsize=(8,6))
modstart=12390
s1.kip_cont(ifig=ifig,ixaxis='age',modstart=modstart,modstop=13240,t0_model=modstart,\
            xres=res,yres=res,engenlevels=engenlevels,\
            mix_zones=20, burn_zones=20,engenPlus=True, engenMinus=False,outlines=True,fsize=fsize)
ylim(0.49,0.56)
#xlabel('$log(t_{final}-t)\ [yr]$',fontsize=fsize,labelpad=1)
savefig('hif_M1_1e-4_kipcont.png',dpi=300)
```
